# 1. Contexto e Objetivos do Projeto
Este projeto tem como objetivo analisar criptomoedas utilizando dados reais da API CoinGecko e agrupá-las em perfis de risco (baixo, médio e alto) por meio de técnicas de Machine Learning Não Supervisionado.

**Objetivo de Negócio:**
Auxiliar investidores a diversificar seus portfólios, identificando grupos de criptomoedas com perfis de risco semelhantes.

**Objetivo de Ciência de Dados:**
Desenvolver uma pipeline completa que envolva:
- Coleta de dados reais via API
- Engenharia de features
- Clusterização (K-Means e PCA)
- Avaliação dos clusters (silhouette score)
- Visualização gráfica dos resultados

O projeto busca agregar valor ao processo de tomada de decisão de investidores, facilitando a identificação de ativos com comportamentos similares em termos de retorno, volatilidade e outros indicadores de risco.

# 2. Pesquisa e Enquadramento do Problema
O projeto foi enquadrado como um problema de **Aprendizagem Não Supervisionada**, utilizando técnicas de clusterização para agrupar criptomoedas conforme seus perfis de risco.

**Perguntas de Negócio:**
- Quais grupos de criptomoedas apresentam comportamentos de risco semelhantes?
- Como identificar ativos com maior potencial de retorno ou maior volatilidade?
- Quais métricas são mais relevantes para diferenciar os perfis de risco?

**Enquadramento do Problema:**
- Tipo: Aprendizagem Não Supervisionada (Clusterização/Redução de Dimensionalidade)
- Algoritmos: K-Means para agrupamento, PCA para visualização
- Métricas: Silhouette Score para avaliação da separação dos clusters

A pesquisa envolveu o levantamento de métodos de agrupamento de ativos financeiros, análise de métricas de risco e retorno, e a definição das variáveis mais relevantes (em nossa visão) para o contexto do projeto.
